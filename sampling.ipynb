    Systematic sampling has a problem: if the data has been sorted, or there is some sort of pattern or meaning behind the row order, then the resulting sample may not be representative of the whole population. The problem can be solved by shuffling the rows, but then systematic sampling is equivalent to simple random sampling.

In general, it is only safe to use systematic sampling if a plot like this has no pattern; that is, it just looks like noise.

### Stratified and weighted random sampling

Stratified sampling allows us to sample a population that contains subgroups. For example, we could group the coffee ratings by country.

```python
top_counted = coffee['country_of_origin'].value_counts().head(6)
top_counted
```

```python
# Filtering for 6 countries
top_counted_subset = coffee.loc[coffee['country_of_origin'].isin(top_counted.index)]
top_counted_subset
```

```python
coffee_samp = top_counted_subset.sample(frac=0.1,random_state=2021)
coffee_samp
```

```python
coffee_samp['country_of_origin'].value_counts(normalize=True)
# Sample
```

```python
top_counted_subset['country_of_origin'].value_counts(normalize=True)
# Population
```

```python
# Proportional stratified sampling
coffee_strat = top_counted_subset.groupby('country_of_origin').sample(frac=0.1, random_state=2021)
coffee_strat['country_of_origin'].value_counts(normalize=True)
```

    By grouping then sampling, the size of each group in the sample is representative of the size of the sample in the population

    Now the proportions of each country in the stratified sample are much closer to those in the population

```python
# Equal counts stratified sampling
coffee_eq = top_counted_subset.groupby('country_of_origin').sample(n=15,random_state=2021)
coffee_eq['country_of_origin'].value_counts(normalize=True)
```

    If one subgroup is larger than another subgroup in the population, but you don't want to reflect that difference in your analysis, then you can use equal counts stratified sampling to generate samples where each subgroup has the same amount of data. For example, if you are analyzing blood types, O is the most common blood type worldwide, but you may wish to have equal amounts of O, A, B, and AB in your sample.



```python
coffee_eq['country_of_origin'].value_counts()
```

# Weighted Random Sampling
***specify weights to adjust the relative probability of a row being sampled***

The probability of picking any given row is proportional to the weight value for that row.

```python
coffee_weight = top_counted_subset
condition = coffee_weight['country_of_origin']=='Taiwan'

coffee_weight['weight'] = np.where(condition,2,1)
# Taiwanese coffees have two times the chance of being picked compared to other coffees
coffee_weight
```

```python
coffee_weight = coffee_weight.sample(frac=0.1,weights='weight')
coffee_weight
```

### Cluster Sampling
One problem with standard sampling is that we need to collect data from every subgroup. In cases where collecting data is expensive, for example, when we have to physically travel to a location to collect it, it can make our analysis prohibitively expensive.

First, you randomly sample which subgroups to include in the sample, then randomly sample rows within each subgroup.

-  Stratified sampling 
    - split the population into subgroups
    - Use simple random sampling on every subgroup
- Cluster sampling 
    - Use simple random sampling to pick some subgroups
    - Use simple random sampling on only those subgroups as before

- Properties:
    - Really noisy, particular for small samples
    - The amplitude of the line is quite steep, to begin with, when we have a small sample size, adding just a few more samples can give us much better accuracy.

### Creating a sampling distribution

```python
mean_cup_point_1000 = []

for i in range(1000):
    mean_cup_point_1000.append(coffee.sample(n=30)['total_cup_points'].mean())
```

```python
import seaborn as sns
sns.histplot(mean_cup_point_1000)
```

```python
plt.hist(mean_cup_point_1000,bins=30)
plt.show()
```

```python
def sampling(size):
    lst = []
    for i in range(1000): # create 1000 replicates
        lst.append(coffee.sample(n=size)['total_cup_points'].mean())
    return lst
```

```python
sizes = [6,30,150]
plt.rcParams["figure.autolayout"] = True
fig,ax = plt.subplots(1,3)
for n, size in enumerate(sizes):
    sns.histplot(sampling(size),bins=30,ax=ax[n])
    ax[n].set_title(f'size:{size}')
    ax[n].set_xlabel('total_cup_points')
    ax[n].axvline(np.array(sampling(size)).mean(),color='red',ls = '--')

```

    As sample size increases, on average each sample mean has a lower relative error compared to the population mean, thus reducing the range of the distribution.

- sampling: going from a population to a smaller sample
- bootstrap: building up a theoretical population from the sample
***
- Bootstrapping use case:
    - Develop understanding of sampling variability using a single sample
***
Bootstrapping process:
1. Make a resample of the same size as the original sample
2. Caluculate a statistic, such as a mean of one of the columns
3. Repeat steps 1 and 2 many times to get lots of these bootstrap statistics

Booststrap distribution mean:
- Usually close to the sample mean
- May not be a good estimate of the population mean
- If the sample is not closely representative of the population, then the mean of the bootstrap distribution will not be representative of the population mean.

***Bootstrapping cannot correct biases from sampling***

Estimated standard error -> standard deviation of the boostrap distribution for a sample statistic

Although bootstrapping was poor at estimating the population mean, it is generally great for estimating the population standard deviation.

Population std.dev $\approx$ Std.Error x $\sqrt{Samplesize}$




    The sampling distribution mean is the best estimate of the true population mean; the bootstrap distribution mean is closest to the original sample mean.

```python
spotify_samp = spotify_population.reset_index().sample(n=5000,replace=False)
spotify_samp
```

```python
spotify_samp.popularity.mean()
```

```python
# create sampling distribution
sampling_distribution =[]
for i in range(5000):
    sampling_distribution.append(
        spotify_population.sample(n=5000)['popularity'].mean()
    )
```

```python
# create boostrap distribution
bootstrap_distribution = []
for i in range(5000):
    bootstrap_distribution.append(
        np.mean(spotify_samp.sample(frac=1,replace=True)['popularity'])
    )
```

```python
spotify_samp
```

```python
# Calculate the population mean popularity
pop_mean = spotify_population['popularity'].mean()

# Calculate the original sample mean popularity
samp_mean = spotify_samp['popularity'].mean()

# Calculate the sampling dist'n estimate of mean popularity
samp_distn_mean = np.mean(sampling_distribution)

# Calculate the bootstrap dist'n estimate of mean popularity
boot_distn_mean = np.mean(bootstrap_distribution)

# Print the means
print([pop_mean, samp_mean, samp_distn_mean, boot_distn_mean])
```

    Conclusions:
    - Sampling distriubtion mean is the best estimate of the true population mean
    - The bootstrap distribution mean is closest to the original sample mean

```python
# Calculate the population std dev popularity
pop_sd = spotify_population['popularity'].std(ddof=0)

# Calculate the original sample std dev popularity
samp_sd = spotify_samp['popularity'].std(ddof=1)

# Calculate the sampling dist'n estimate of std dev popularity
samp_distn_sd = np.std(sampling_distribution, ddof=1) * np.sqrt(len(spotify_samp))

# Calculate the bootstrap dist'n estimate of std dev popularity
boot_distn_sd = np.std(bootstrap_distribution, ddof=1) * np.sqrt(len(spotify_samp))

# Print the standard deviations
print([pop_sd, samp_sd, samp_distn_sd, boot_distn_sd])
```

    Conclusion:
    This is an important property of the bootstrap distribution. When you don't have all the values from the population or the ability to sample multiple times, you can use bootstrapping to get a good estimate of the population standard deviation.

### Confidence interval

```python
coffee_samp = coffee.sample(frac=0.4,replace=False)
bootstrap_dist_flavor = []
for i in range(2000):
    bootstrap_dist_flavor.append(
        np.mean(coffee_samp.sample(frac=1,replace=True)['flavor'])
    )
plt.hist(bootstrap_dist_flavor,bins=15)
plt.show()
```

```python
# Mean of the resample
mean_resamp = np.mean(bootstrap_dist_flavor)
mean_resamp
```

```python
# Mean plus or minus on standard deviation
lower_bond = np.mean(bootstrap_dist_flavor) - np.std(bootstrap_dist_flavor,ddof=1)
lower_bond
```

```python
upper_bond = np.mean(bootstrap_dist_flavor) + np.std(bootstrap_dist_flavor,ddof=1)
upper_bond
```

```python
plt.hist(bootstrap_dist_flavor,bins=15)
plt.axvline(lower_bond,ls='--',color='red')
plt.axvline(upper_bond,ls='--',color='red')
plt.axvline(mean_resamp,ls='--',color='black')
plt.show()
```

### Quantile method for confidence intervals

```python
np.quantile(bootstrap_dist_flavor,0.025)
```

```python
np.quantile(bootstrap_dist_flavor,0.975)
```

### Standard error method for confidence interval

```python
point_estimate = np.mean(bootstrap_dist_flavor)
point_estimate
```

```python
std_error = np.std(bootstrap_dist_flavor,ddof=1)
std_error
```

The standard error method for calculating the confidence interval assumes that the bootstrap distribution is normal. This assumption should hold if the sample size and number of replicates are sufficiently large.

```python
from scipy.stats import norm
lower_se = norm.ppf(0.025,loc=point_estimate,scale=std_error)
upper_se = norm.ppf(0.975,loc=point_estimate,scale=std_error)
print((lower_se,upper_se))
```

    Confidence intervals account for uncertainty in our estimate of a population parameter by providing a range of possible values. We are confident that the true value lies somewhere in the interval specified by that range.
